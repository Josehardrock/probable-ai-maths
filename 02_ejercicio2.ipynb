Ejercicio 2.
Usando el método del Descenso del Gradiente explore el comportamiento y la eficacia del
algoritmo en las siguientes funciones y su visualización.
1. ℎ(x) = cos(x) + x²
2. ℎ(x) = e -^x (x² + 5x)
3. ℎ(x) = ln(x + 1) + x²
4. ℎ(x) = x^5 − 5x³ + 4x
Instrucciones:
1. Derivación:
a. Derive cada función manualmente o con herramientas computacionales para
encontrar su gradiente.
2. Implementación del Algoritmo:
a. Codifique una función que implemente el algoritmo de Descenso del Gradiente.
