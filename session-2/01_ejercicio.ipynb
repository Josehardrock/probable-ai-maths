Instrucciones:

2. Aplicación del Algoritmo:<br>

```python
# Function del descenso del gradiente
def gradient_descent(func, d_func, x_start=1.0, lr=0.01, max_iters=100) :

    """
    Function to eval the gradient descent algorithm. 
    :param func: function to minimize
    :param d_func: derivative of the function
    :param x_start: starting point
    :param lr: learning rate
    :param max_iters: maximum number of iterations
    """

    x=sp.Symbol('x')
    # Add the first point to the trajectory
    trajectory = [x_start]
    x = x_start
    
    for i in range(max_iters):
        gradient = d_func(x)
        x = x - lr * gradient

        trajectory.append(x)

    print(f"LastPoint : {x} \nTrajectory : {trajectory[:5]} ... {trajectory[-5:]}, ")
    return trajectory, x
```

```python
# Plot de la funcion
def plot_gradient_descent(func, trajectory, last_point_f, title="") :
    """
    Function to plot the gradient descent algorithm
    :param func: function to plot
    :param trajectory: trajectory of the algorithm
    :param point: last point
    """
    # Crear un rango de valores x para evaluar la función
    x_values = np.linspace(min(trajectory) - 1, max(trajectory) + 1, 100)
    y_values = func(x_values)

    # Convertir la trayectoria en arrays para graficar (Me tomó tiempo entender esto)
    trajectory = np.array(trajectory)
    y_trajectory = func(trajectory)

    # Graficar la función
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=f"Function {title}", color="blue")
    plt.scatter(trajectory, y_trajectory, color="red", label="Trajectory", zorder=5)
    plt.plot(trajectory, y_trajectory, color="red", linestyle="--", alpha=0.7) # Pintar la trayectoria de la función

    x_start, y_start = trajectory[0], y_trajectory[0]
    start_label =f"Starting point ( {x_start}, {y_start} )"
    x_last, y_last = trajectory[-1], y_trajectory[-1]
    last_label = f"Last point ( {x_last}, {y_last} )"

    # Marcar puntos inicial y final
    plt.scatter(x_start, y_start, color="green", label=start_label, zorder=6)
    plt.scatter(x_last, y_last, color="purple", label=last_label, zorder=6)

    # Encontrar el minimo de la función
    min_result = minimize(func, x0=1, bounds=[(-1, 3)])
    plt.scatter(min_result.x, min_result.fun, color='black', label=f"Minimum of f'(x) [-1,3] {min_result.fun}")

    # Configurar etiquetas y título
    title = f"Gradient Descent - {title}"
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid()
    plt.show()
```

$$
 a)  f(x) = x^3 + 3x^2 - 2, f'(x) = 3x^2 + 6x

```python
def f(x) : return x**3 + 3*x**2 -2
def f_prime(x) : return 3*x**2 + 6*x

trajectory_f, last_point_f = gradient_descent(f, f_prime, x_start=1.0, lr=0.025, max_iters=100)
plot_gradient_descent(f, trajectory_f, last_point_f, title="f(x) = x^3 + 3x^2 - 2")
```
