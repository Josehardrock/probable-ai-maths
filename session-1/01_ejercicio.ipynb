```python
# Álgebra Lineal
# Análisis de Preferencia de Productos
# Supongamos que estás trabajando para una tienda online y quieres entender las preferencias de
# tus clientes en relación con cuatro categorías de productos: tecnología, moda, libros y deportes.
# Para ello, has extraído en una matriz el número de compras para cada categoría en un periodo
# determinado:

M = [
        [4, 1, 6, 2],
        [2, 8, 1, 3],
        [5, 2, 3, 9]
        ]

# donde las filas representan los diferentes clientes y las columnas representan las categorías
# respectivamente.
# Empleando esta matriz, desarrolle un algoritmo en python que:
# a) Calcule el total de compras por categoría
# b) Identifique que cliente ha comprado más unidades en total
# c) Obtenga el gasto por cliente empleando la siguiente matriz de precios por categoría 

M = [
90,
150,
50,
110,
]
```

```python
import numpy as np

cust_purchases = np.array(
                [
                [4, 1, 6, 2], 
                [2, 8, 1, 3], 
                [5, 2, 3, 9] 
                ])

prices = np.array([90, 150, 50, 110])
print(f"Purchases {cust_purchases}")
print(f"Prices {prices}")

# a) Calcule el total de compras por categoría

def by_category(purchases):
    return purchases.sum(axis=0)

def by_customer(purchases):
    return purchases.sum(axis=1)

total_purchases = cust_purchases.dot(prices)
print(f"a) Total de compras por categoría: {by_category(cust_purchases)}")
print(f"b) Cliente que ha comprado más unidades en total: {np.argmax(by_customer(cust_purchases))}")

def customer_expense(purchases, prices):
    return purchases.dot(prices)

print(f"Calc Total de compras por categoría {customer_expense(cust_purchases, prices)}")
```
